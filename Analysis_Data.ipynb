```python
# Train  data set
#Here we will check the precentage of nan values present in each feature
#List the missing values in each featureabs
features_with_nan = [feature for feature in data.columns if data[feature].isnull().sum()>1]
#print the feature name and the precentage of missing values

for feature in features_with_nan:
    print(feature, np.round(data[feature].isnull().mean(),4),'% missing values')
```

# __Missing Values__

```python
# Train  data set
#Let's us capture all the nan values
# let's handle ccategorical_features which are missing
c_feature_nan_train = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtype=='O']

for feature in c_feature_nan_train:
    print('{}: {}% missing values'.format(feature,np.round(data[feature].isnull().mean(),4)))
```

```python
# Test data set
# let's handle categorical_features which are missing
c_feature_nan_test = [feature for feature in test_df.columns if test_df[feature].isnull().sum()>1 and test_df[feature].dtype=='O']

for feature in c_feature_nan_test:
    print('{}: {}% missing values'.format(feature,np.round(test_df[feature].isnull().mean(),4)))
```

```python
# Train data set
#Replace missing value with a new label for train data set
def replace_cat_feature(data,c_feature_nan_train):
    data1=data.copy()
    data1[c_feature_nan_train]=data1[c_feature_nan_train].fillna('Missing')
    return data1

data=replace_cat_feature(data,c_feature_nan_train)

data[c_feature_nan_train].isnull().sum()
```

```python
# Test data set
#Replace missing value with a new label for test data set
def replace_cat_feature(test_df,c_feature_nan_test ):
    data1=test_df.copy()
    data1[c_feature_nan_test]=data1[c_feature_nan_test].fillna('Missing')
    return data1

test_df=replace_cat_feature(test_df,c_feature_nan_test)

test_df[c_feature_nan_test].isnull().sum()
```

```python
# Train data set
# Let's check for numerical_features the contains missing values
n_feature_nan_train = [feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes!='O']

# priting numerical_features and percentage of missing values

for feature in n_feature_nan_train:
    print('{}: {}% Missing value'.format(feature,np.around(data[feature].isnull().mean(),4)))
```

```python
# Test data set
# Let's check for nnumerical_features the contains missing values
n_feature_nan_test = [feature for feature in test_df.columns if test_df[feature].isnull().sum()>1 and test_df[feature].dtypes!='O']

# priting numerical_features and percentage of missing values

for feature in n_feature_nan_train:
    print('{}: {}% Missing value'.format(feature,np.around(test_df[feature].isnull().mean(),4)))
```

```python
# Train dataset
# Replacing the numerical missing values 

for feature in n_feature_nan_train:
    # we will replace by using median beacuse of outlier present 
    median_value = data[feature].median() 

    # create a new feature to capture nan values
    data[feature+'nan']=np.where(data[feature].isnull(),1,0)
    data[feature].fillna(median_value,inplace=True)

data[n_feature_nan_train].isnull().sum()
```

```python
# Test dataset
# Replacing the numerical missing values 

for feature in n_feature_nan_test:
    # we will replace by using median beacuse of outlier present 
    median_value = test_df[feature].median() 

    # create a new feature to capture nan values
    test_df[feature+'nan']=np.where(test_df[feature].isnull(),1,0)
    test_df[feature].fillna(median_value,inplace=True)

test_df[n_feature_nan_train].isnull().sum()
```

In above we created a new feature with nan column name and stored the values 1 when the coulmn having nan value and 0 when the column doesn't have nan values

```python
# Train data set 
# Temporal variable (Data Time variables)

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    data[feature]=data['YrSold']-data[feature]
```

# __Handling rare categorical feature__

We will remove categorical vairables the are present less than 1% of the observations

```python
# Train data set
for feature in categorical_features_train:
    temp = data.groupby(feature)['SalePrice'].count()/len(data)
    temp_df = temp[temp>0.01].index
    data[feature]=np.where(data[feature].isin(temp_df),data[feature],'Rere_var')
```

```python
# Test data set
for feature in categorical_features_test:
    temp = test_df.groupby(feature).count()/len(data)
    temp_df = temp[temp>0.01].index
    test_df[feature]=np.where(test_df[feature].isin(temp_df),test_df[feature],'Rere_var')
```

```python
# Test data set 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in categorical_features_test:
                  test_df[i] = test_df[i]=le.fit_transform(test_df[i])
```
