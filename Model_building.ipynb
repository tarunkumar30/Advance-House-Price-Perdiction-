# __Feature Engineering__ 

 #### __Handling missing values__ 

```python
# Train data set 
# check null value are present in data set 
train_data.isnull().sum().sum()
```

```python
# Test data set 
# check null value are present in data set
null_columns= test_data.columns[test_data.isnull().any()]
print(test_data[null_columns].isnull().sum())
```

```python
# Test data set 
test_data[null_columns]=test_data[null_columns].fillna(test_data[null_columns].median())
```

```python
# Test data set 
# after filling null values with median 
test_data.isnull().sum().sum()
```

I use median to fill the null values beacuse median well perfrom on outliers so i choose to fill null values with median 

#### __Handling outliers__

```python
#Train data set
# out liear removing using z score 
outlier_col_train=[i for i in train_data.columns]
z_score = (train_data[outlier_col_train]-train_data[outlier_col_train].mean())/(train_data[outlier_col_train].std())

# Define tthershold value
thereshold = 3

# Create a mask: keep rows where all selected features have a z-score <3
keep_mask = (z_score.abs()<thereshold).all(axis=1)

# save the filtered data back to train
train_data = train_data[keep_mask]

print(f'new  train dataaset shape:{train_data.shape}')
```

I found that having more outliers in our data set 

if i add those outliers for my model mans my model get effect from those outliers and reduce my model score so i drop it 

```python
# Test data set
# out liear removing using z score 
outlier_col_test=[i for i in test_data.columns]
z_score = (test_data[outlier_col_test]-test_data[outlier_col_test].mean())/(test_data[outlier_col_test].std())

# Define tthershold value
thereshold = 3

# Create a mask: keep rows where all selected features have a z-score <3
keep_mask_test = (z_score.abs()<thereshold).all(axis=1)

# save the filtered data back to train
test_data = test_data[keep_mask_test]

print(f'new test dataaset shape:{test_data.shape}')
```

 # __Feature selection__

```python
# Define x and y for the taining set 
x_train_full = train_data.drop(['Id','SalePrice'],axis=1)
y_train = train_data['SalePrice']
```

```python
# for feature selection  
from sklearn.linear_model import Lasso 
from sklearn.feature_selection import SelectFromModel 
```

```python

## Apply feature selection  
# 1. specify the lasso regression model  
# 2. select a suitable alpha (equialent of penalty) 
# 3. the bigger the alpha the less feature that will be selected  
# 4. then i use the selection model object from sklearn 
# 5. will select the features which coefficients are non-zero  
feature_sel_model_train = SelectFromModel(Lasso(alpha=0.005, random_state=0))  
feature_sel_model_train.fit(x_train_full,y_train) 
```

```python
# create the list of selected columns name
selected_feat = x_train_full.columns[(feature_sel_model_train.get_support())]
```

```python
# Filter both dataset using that list 
# Now both will have exactly 33 columns 
x_train_final = x_train_full[selected_feat]
x_test_final = test_data[selected_feat]
```

```python
feature_sel_model_train.get_support()
```

Those true and false are the count of the independent variables 


True - indicate the improtant of the feature  


False - indicate the columns are not that much improtant 

```python
# Train data set
# print the number of total featurea are selected  
# List for the selected features  
selected_features= np.sum(feature_sel_model_train.get_support())

# print some stats  
print('Selected features are :{}'.format(selected_features))
print('features with coefficients shrank to zero:{}'.format(np.sum(feature_sel_model_train.estimator_.coef_==0))) 
```

In the above we have done feature selection and we selected 41 features from train and test data set for the build model

# __Model__

```python
# Spliting data into train and test 
from sklearn.model_selection import train_test_split
x_train_v,x_val,y_train_v,y_val = train_test_split(x_train_final,y_train,test_size=0.2,random_state=42)
```

```python
# BUild and train the Liner Regression Model 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train_v,y_train_v)
```

```python
# predict on the test set 
predictions = model.predict(x_val)
```

```python
model_accuracy = model.score(x_val,y_val)
print(f'Model R-squred score:{model_accuracy}')
```
